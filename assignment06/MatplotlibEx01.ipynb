## Line plot of sunspot data

Download the `.txt` data for the "Yearly mean total sunspot number [1700 - now]" from the [SILSO](http://www.sidc.be/silso/datafiles) website. Upload the file to the same directory as this notebook.

```python
import os
assert os.path.isfile('yearssn.dat')
```

Use `np.loadtxt` to read the data into a NumPy array called `data`. Then create two new 1d NumPy arrays named `years` and `ssc` that have the sequence of year and sunspot counts.

```python
years,ssc = np.loadtxt('yearssn.dat', usecols=(0,1), unpack=True)
```

```python
assert len(years)==316
assert years.dtype==np.dtype(float)
assert len(ssc)==316
assert ssc.dtype==np.dtype(float)
```

Make a line plot showing the sunspot count as a function of year.

* Customize your plot to follow Tufte's principles of visualizations.
* Adjust the aspect ratio/size so that the steepest slope in your plot is *approximately* 1.
* Customize the box, grid, spines and ticks to match the requirements of this data.

```python
plt.figure(figsize=(16,2))
plt.plot(years,ssc,linestyle='-', marker='o', markeredgecolor = "darkorange",  color='darkorange')
plt.xlabel("$Year$",size = 20)
plt.ylabel("$Sun$$\:$$Spots$",size = 20)
plt.xlim(1700, 2016)
plt.ylim(0, 285)
plt.tick_params(axis='y', direction='out', length=5, labelright = True);
```

Describe the choices you have made in building this visualization and how they make it effective.

##### I chose to color the line orange so that it would bring the eye to the data. 
##### I chose to leave the labels the default color so that the viewer is more inclined focus on the data.
##### I enlarged the size of the labels so that it was easier for the viewer to read.
##### I made the y-ticks point outwards so as to not distract the viewer form the data.
##### I made the markers dark-orange circles to make them look like sun spots.

Now make 4 subplots, one for each century in the data set. This approach works well for this dataset as it allows you to maintain mild slopes while limiting the overall width of the visualization. Perform similar customizations as above:

* Customize your plot to follow Tufte's principles of visualizations.
* Adjust the aspect ratio/size so that the steepest slope in your plot is *approximately* 1.
* Customize the box, grid, spines and ticks to match the requirements of this data.

```python
f, ax = plt.subplots(2, 2, figsize=(12,6))
for i in range(2):
    for j in range(2):
        plt.sca(ax[i,j])
        coef = (2*i+j)
        if coef == 3:
            k = 315
        else:
            k = 100*(coef + 1)
        year_End = (coef+1)*100 + 1700
        year_Start = coef*100 + 1700
        plt.plot(years[coef*100:k],ssc[coef*100:k], linestyle='--', marker='D', \
                 markeredgecolor = "darkorange",  color='darkorange')
        plt.xlim(year_Start, year_End)
        plt.ylim(0, 285)
        if i == 1:
            plt.xlabel("$Year$")
        if j == 0:
            plt.ylabel("$Sun$$\:$$Spots$")
        
        plt.tick_params(axis='y', direction='out', length=5);
plt.tight_layout()
```
