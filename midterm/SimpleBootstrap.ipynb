# Simple Bootstrap

In this notebook you are going to draw samples from a probability distribution of your choice, and then compute a bootstrapped mean and its confidence interval.

## Imports

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

## Boostrapping the mean

Perform the following steps:

* Draw `100` samples from a continuous probability distribution of your choice with a given mean and variance.
* Bootstrap your raw data `500` times and use the bootstrapped data to compute the distribution of the mean of the samples.
* Make an appropriately labeled visualization of the distribution of the mean.
* Compute and print the 90% confidence interval of the mean.

```python
n = 10000000
N = 5000
mu = 100
var = 15**2
data = np.random.normal(mu,var**.5,n)
bootstrap = np.empty(N)
for j in range(N):
    bootstrap[j] = (np.random.choice(data,1,replace=True))
```

To Confirm the validity of the probability distribution:
   
   The probability of 172 occurring in a population of 10,000,000 should be tiny 

```python
def calc_prob(data,val):
    x = 0
    n = len(data)
    for i in range(n):
        if data[i] > val:
            print(data[i])
            x += 1
    print(x)
calc_prob(data,172)
```

```python
sns.distplot(bootstrap)
plt.title("$E[N(100,15)]$", size = 20)
plt.xlabel("$Mean$", size = 20)
plt.ylabel("$Count$", size = 20);
```

```python
print("90% Confidence Interval for the Mean: (",np.percentile(bootstrap,5),",",np.percentile(bootstrap,95),")")
```
